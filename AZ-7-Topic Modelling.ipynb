```
#We do not know what topic these articles belong to
#We try to figure the topic for each of these Articles
#Before running LDA, we perform a little bit of preprocessing
#We perform Count Vectorization
from sklearn.feature_extraction.text import CountVectorizer
```

```
#Create an instance of Count Vectorizer
#We ignore certain terms for example common words.
#So, we specify maximum and minimum frequency for a term to be included in the document term matrix
#max_df gets rid of terms that are really common across the documents
#Value of max_df is between 0 and 1.
#PARAMETER 1:         max_df=0.9 is going to discard words that show up in 90 percent of documents
#PARAMETER 2:         min_df=  is the wrds that show up minimum amount of times
#You can actually pass the values to max_df and min_df as a ratio between 0 and 1
#You can also pass an integer,it will actually take into account as a raw number of documents
#If I pass min_df=2: this says that the minimum frequency for a word to be counted into the Count Vectorizer
#has to show up in atleast two documents
#Benefits achieved due to specifying max_df and min_df
#######This helps in throwing out some common words
#######This helps in throwing out some random words maybe typos and misspelling
#PARAMETER 3:         stop_words:    Specifies the language of stopwords
cv=CountVectorizer(max_df=0.9,min_df=2,stop_words='english')
```

```
#Other methods of performing above
#Tokenize the words
#Remove stopwords using Spacy
```

```
#There is no train test split involved as this is unseupervised
#We are fit transforming to the entire dataset: Creating the Document Term Matrix of the Article Column
dtm=cv.fit_transform(npr['Article'])
```

```
#Check the Document Term Matrix
dtm
```

```
#We have 54777 terms in the document term matrix
#Now perform LDA
from sklearn.decomposition import LatentDirichletAllocation
```

```
#There is no specific number for n_components
#It purely depends on domain experience and familiraty with the domain from where the dataset is from
#If we put n_components=7, taht means we want 7 topics to be returned
#If we want more details like... National Politics,State Politics under Politics topic
#You can increase the n_components
LDA=LatentDirichletAllocation(n_components=7,random_state=42)
```

```
#Fit LDA to document term matrix
LDA.fit(dtm)
#This will take time
```

```
#PART 3--------------------------#Grab the highest probability words per topic
#Let us take one topic
#We are considering the very first topic
#We still do not know what this topic actually represents
single_topic=LDA.components_[0]
```

```
#Example to explain argsort
import numpy as np
```

```
arr=np.array([10,200,1])
```

```
arr
```

```
arr.argsort()
#The index positions are from lowest to highest
#Here index 2 value is 1 which is smallest
#Here index 0 value is 10
#Here index 1 value is 200 which is largest 
```

```
#argsort returns index positions that would sort this array
single_topic.argsort()
#We are taking single topic and then figure out what index positions we should be looking at
#for high probability words for this particular single topic
```

```
#Now let us grab the top ten words for this topic
#ARGSORT----> returns index positions sorted from LEAST to GREATEST
#We are looking for TOP 10 values. SO we need the last ten values of argsort
#Below statement grabs the last ten values of .argsort
single_topic.argsort()[-10:]
```

```
#Let us try to indntify those top ten words with the help of indexes obtained in the above step
top_ten_words=single_topic.argsort()[-10:]
```

```
for index in top_ten_words:
  print(cv.get_feature_names()[index])
```

```
#Now I want to take a look at top 20 words. This helps in understanding the topic more
top_twenty_words=single_topic.argsort()[-20:]
```

```
for index in top_twenty_words:
  print(cv.get_feature_names()[index])
#Looks like it is realted to some Health Insurance and policy
```

```
#Let us create a loop that prints out top 15 words for each of the 7 topics
for i,topic in enumerate(LDA.components_):
  print(f"The top 15 words for the TOPIC #{i}")
  print([cv.get_feature_names()[index] for index in topic.argsort()[-15:]])
  print('\n')
  print('\n')
```
