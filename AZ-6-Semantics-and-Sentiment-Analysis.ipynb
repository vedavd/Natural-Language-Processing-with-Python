```
#Calculate new vectors by adding or subtracting vectors
#king-man=Queen
#We gotta calculate cosine similarities 
from scipy import spatial
```

```
cosine_similarity=lambda vec1,vec2: 1- spatial.distance.cosine(vec1,vec2)
```

```
king=nlp.vocab['king'].vector
man=nlp.vocab['man'].vector
woman=nlp.vocab['woman'].vector
```

```
#King-man+woman--------> NEW_VECTOR similar to Queen,princess,highness
new_vector=king-man+woman
```

```
computed_similarities=[]

#I am going through the entire vocabulary(684 thousand words) and find out words from the vocabulary that are similar to the new_vector
for word in nlp.vocab:
  if word.has_vector:
    if word.is_lower:
      if word.is_alpha:
        similarity=cosine_similarity(new_vector,word.vector)
        computed_similarities.append((word,similarity))
```

```
#Let us check the computed_similarities
computed_similarities
```

```
#We are arranging in descending order
#-item[1] will sort in descending  order
computed_similarities=sorted(computed_similarities,key=lambda item:-item[1])
```

```
#Let us check the words in the computed_similarities
print([t[0].text  for t in computed_similarities])
```

```
#Let us check the top ten words in the computed_similarities
print([t[0].text for t in computed_similarities[:10]])
```

```
#Let us analyze amazon reviews
import pandas as pd
```

```
df=pd.read_csv('/content/amazonreviews.tsv',sep='\t')
```

```
df.head()
```

```
#Get the count of positive and negative labels
df['label'].value_counts()
```

```
#Check if there are any null values
df.isnull().sum()
```

```
#However let us drop the na
df.dropna(inplace=True)
```

```
#Check for white spaces and then remove
blanks=[]
for i,lb,rv in df.itertuples():
  if type(rv) == str:
    if rv.isspace():
      blanks.append(i)
```

```
#Check if anything is appended to blanks
blanks
#So there are no white spaces at all
```

```
#BUt we are going to drop whatever is present in blanks
df.drop(blanks,inplace=True)
```

```
#Check the length of the dataframe
#As there are no blank reviews and white  spaces
#The current length of data frame should be equal to the original one
len(df)
```

```
#Let us take a look at the first review that is positive
df['review'][0]
```

```
#the polarity scores for the above should show as positive
sid.polarity_scores(df.iloc[0]['review'])
```

```
#Now let us add scores of reviews to the dataframe
df['scores']=df['review'].apply(lambda review:sid.polarity_scores(review))
```

```
#A new column should be added with the scores
df.head()
```

```
#We want the compound score alone to be added to a column
df['compound']=df['scores'].apply(lambda d:d ['compound'])
```

```
df.head()
```

```
#Based on the compound score
#if the compound score is greater than 0, then it is positive review
#if the compound score is less than zero then itis negative review
df['comp_score']=df['compound'].apply(lambda score:'pos' if score>=0 else 'neg')
```

```
df.head()
```

```
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```
accuracy_score(df['label'],df['comp_score'])
```

```
print(classification_report(df['label'],df['comp_score']))
```

```
dd=pd.DataFrame(confusion_matrix(df['label'],df['comp_score']),index=['positive','negative'],columns=['positive','negative'])
```
