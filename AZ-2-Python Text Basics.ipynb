```python
#This is a tuple
#Assume one of the title is so long
library=[('Author','Topic','Pages'),('Twain','Rafting in water alone',601),('Feynman','Physics',95),('Hamilton','Mythology',144)]
```

```python
library
```

```python
lenau=0
lentop=0
for author,topic,pages in library:
    x=len(author)
    if x>lenau:
        lenau=x
    y=len(topic)
    if y>lentop:
        lentop=y
print(f"The maximum length of an Author is {lenau} characters")
print(f"The maximum length of a Topic is {lentop} characters")
```

```python
#Indentation
#We are arranging taken into consideration the maximum length of characters
for author,topic,pages in library:
    print(f"{author:{lenau+2}} {topic:{lentop+2}} {pages}")
```

```python
#Indentation
#We are arranging by just passing whitespaces into the columns
for author,topic,pages in library:
    print(f"{author:{10}} {topic:{30}} {pages:{10}}")
```

```python
#Here pages is not aligned
#So, we just include > symbol in the indentation provided for pages column
for author,topic,pages in library:
    print(f"{author:{10}} {topic:{30}} {pages:>{10}}")
```

```python
#We can add any pattern that you wanna pad with
#Here we set pages column length as 10 characters
#Word pages consumes only 5 characters.
#Remaining 5 characters can be padded with any pattern
for author,topic,pages in library:
    print(f"{author:{10}} {topic:{30}} {pages:*>{10}}")
```

```python
#We cannot write a string to the PDF
#But we can copy pages and append pages
#You can copy page from anotehr pdf
#You can copy page from same pdf
#Append it to the current pdf you are working with
#What you cannot do is take a python string and insert into somewhere in the PDF
```

```python
#Let us take a look at the new pdf containing the added pdf page
#The MY_BRAND_NEW pdf file is currently not existing
#But if we open a file with write permission, it will automatically create one
pdf_output=open('MY_BRAND_NEW.pdf','wb')
```

```python
pdf_writer.write(pdf_output)
```

```python
#I am closing the output file
pdf_output.close()
```

```python
#I am closing the original file as well
f.close()
```

```python
#Grab all the text from the entire PDF file
f = open('US_Declaration.pdf','rb')

#create an empty list
pdf_text=[]

#Create a pdf reader object
pdf_reader=PyPDF2.PdfFileReader(f)


for p in range(pdf_reader.numPages):
    
    page=pdf_reader.getPage(p)
    
    #Extract text from the page
    page.extractText()
    
    #Append the above to the blank list created initially
    pdf_text.append(page.extractText())


#Close the file
f.close()
```

```python
#Let us view the entire text of the read PDF file
pdf_text
```

```python
#Length of list that is containing read text from the entire pdf
len(pdf_text)
```

```python
for page in pdf_text:
    print(page)
    print('\n')
    print('\n')
    print('\n')
    print('\n')
    print('\n')
```

# 5. Regular Expressions Part One

```python
text="The phone number of the agent is 123-123-1234.Call soon!"
```

```python
#Check if word phone is present in the text
"phone" in text
```

```python
#We wanna look for a phone number in the text
#import regular expression library
import re
```

```python
pattern="phone"
```

```python
re.search(pattern,text)
```

```python
my_match=re.search(pattern,text)
```

```python
my_match.span()
#The pattern is starting at index 4
#The pattern is ending at index 9 meaning position 8
```

```python
#Get the starting index of the match
my_match.start()
```

```python
#Get the ending index of the match
my_match.end()
```

```python
#if the pattern is present more than once
text="my phone is a new phone"
```

```python
match=re.search(pattern,text)
```

```python
#The span will give start and end index of the first match only
#It will not give the start and end indexes of the subsequent matches
match.span()
```

```python
#Here pattern is present in following indexes
#(3,8)
#(18,23)
#If we want to find the indexes of all the matches
all_matches=re.findall(pattern,text)
```

```python
#Check how many times the pattern is present in the text
len(all_matches)
#Here phone is present twice
```

```python
#The following gives the start and end indexes of all the matches
for match in re.finditer(pattern,text):
    print(match.span())
```
